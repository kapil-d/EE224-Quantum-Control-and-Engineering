```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.signal import chirp, find_peaks, peak_widths
import math
```

```python

def filterfunc(x, N):
    return 16*(np.sin(0.5*np.pi*x)**4)*(np.sin(np.pi*N*x)**2) / (4*(np.pi**2)*(x**2)
    *(np.cos(np.pi*x)**2))
```

```python
x = np.linspace(0,1,1000)
filter = filterfunc(x, 20)
peaks, _ = find_peaks(filter)
results_half = peak_widths(filter, peaks, rel_height=0.5)
plt.plot(filter)
plt.plot(peaks, filter[peaks], "x")
plt.hlines(*results_half[1:], color="C2")
plt.show
```
